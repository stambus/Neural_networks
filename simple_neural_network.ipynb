```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = keras.datasets.fashion_mnist
```

```python
(train_images, train_labels), (test_images, test_labels) = data.load_data()
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
train_images = train_images / 255.0
test_images = test_images / 255.0
```

```python
my_model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(128, activation="relu"),
    keras.layers.Dense(10, activation="softmax")
])
```

```python
my_model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=['accuracy'])
```

```python
my_model.fit(train_images, train_labels, epochs=5)
```

```python
test_loss, test_acc = my_model.evaluate(test_images, test_labels)
```

```python
prediction = my_model.predict(test_images)
class_names[np.argmax(prediction[0])]
```

```python
for i in range(5):
    plt.grid(False)
    plt.imshow(test_images[i], cmap=plt.cm.binary)
    plt.xlabel("Actual: " + class_names[test_labels[i]])
    plt.title("Predictioon " + class_names[np.argmax(prediction[i])])
    plt.show()
```
